## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Categorical.csv'</i>.

```python
raw_data = pd.read_csv('Categorical.csv')
raw_data.head()
```

## Map the data

Use the <i>'continent'</i> category for this analysis.

```python
raw_data['continent'].unique()
```

```python
len(raw_data['continent'].unique())
```

```python
data = raw_data.copy()
data['continent'] = data['continent'].map({'North America':0,'Asia':1,'Africa':2,'Europe':3,'South America':4,'Oceania':5,'Antarctica':6, 'Seven seas (open ocean)':7})
data.head()
```

## Select the features

```python
data_cont_only = data[['continent']]
data_cont_only.head()
```

## Clustering - 4 clusters

Use 4 clusters initially.

```python
kmeans = KMeans(4)
pred_results = kmeans.fit_predict(data_cont_only)
```

```python
plt.scatter(data['Longitude'],data['Latitude'],c=pred_results,cmap='rainbow')
plt.xlabel='Longitude'
plt.ylabel='Latitude'
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```
