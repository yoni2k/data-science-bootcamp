## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries_exercise.csv'</i>.

```python
raw_data = pd.read_csv('Countries-exercise.csv')
raw_data.head()
```

Remove the duplicate index column from the dataset. 

```python
data = raw_data.drop('name',axis=1)
data.head()
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.title('Countries')
plt.show()
```

## Clustering

Use 4 clusters initially.

```python
kmeans = KMeans(4)
raw_data['Cluster'] = kmeans.fit_predict(data)
raw_data.head()
```

### Clustering results

```python
plt.scatter(raw_data['Longitude'],raw_data['Latitude'],c=raw_data['Cluster'],cmap='rainbow')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.title('Countries')
plt.show()
```

## Selecting the number of clusters

### WCSS and the Elbow Method

Use the ingerated <i>sklearn</i> method <i> 'inertia_' </i>.

```python
kmeans.inertia_
```

Write a loop that calculates and saves the WCSS for any number of clusters from 1 up to 10 (or more if you wish).

```python
inertias = []
for i in range(1, 10):
    inertias.append(KMeans(i).fit(data).inertia_)
inertias
```

```python
plt.plot(range(1,10),inertias)
plt.show()
```

<b>Conclusion</b>: better to use 3 or 5 clusters
